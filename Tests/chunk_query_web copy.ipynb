```python
import purecpp_libs as libs
from purecpp_extract import WebLoader

import purecpp_chunks_clean as cc
```

```python
import faiss
import numpy as np
```

```python
loader = WebLoader("https://www.espn.com/")
docs = loader.Load()
```

```python
chunk_default = cc.ChunkDefault(chunk_size=420, overlap=60)
chunked_docs = chunk_default.ProcessDocuments(items=docs)
# chunked_docs
```

```python
q_chunker = cc.ChunkQuery()
```

```python
# print(q_chunker)
```

```python
query = q_chunker.Query("Playoffs of")
# query["embeddings"] = query["embeddings"].astype("float32")
print(query)
```

```python
# query["embeddings"] = query["embeddings"].astype("float32")
qe = np.array(query.embedding, dtype=np.float32).reshape(1, -1)
```

```python
embeddings = q_chunker.CreateVD(chunked_docs)
embeddings
```

```python
array_np = np.array(embeddings, dtype=np.float32)
#dim = embeddings.shape[1] #numpay array
dim = len(embeddings[0])  
dim
```

```python
index = faiss.IndexFlatL2(dim)
index.add(array_np)
```

```python
distances, indices = index.search(qe, 2)

print("Indices retornados pelo FAISS:", indices)
print("Distâncias:", distances)
```

```python
print("### Question:")
print(query.page_content)
print()
print("### Relevant context:")
for i, idx in enumerate(indices[0]):
    print(f"{i+1}. Score [{distances[0][i]:.4f}]: {chunked_docs[idx].page_content}")
```
